NumPy ***ndarray*** is a homogenous n-dimensional array object

A python list or pandas DataFrame can contain:
- a mix of strings
- numbers
- objects (mix of different data types)

Homogenous means all data have to have the ***same*** data type, for ex: all have to be floating point number

- single column (1-dimensional) 
- to the matrix (2-dimensional)

# Create an array with 2 dimension (matrix):

```python
array_2d = np.array([[1, 2, 3, 9], [5, 6, 7, 8]])
```

We have 2 pairs of square brackets. This array has 2 rows and 4 columns

NumPy refers to dimensions as ***axes***
- the first axis has length 2
- the second axis has length 4

You can access a particular row or value with the square bracket notation.
- To access a particular value, provide an index for each dimension
- we have two dimension, so we have to provide index for the row and column

```python
array_2d[1, 2]
```

To access an entire row and all the values therein, we can use the : operator

```python
array_2d[0, :]
```

# Broadcasting, Scalars and Matrix Multiplication


<h2> Linear Algebra with Vectors </h2>

NumPy is designed to do math (and do it well!). This means that NumPy will treat vectors, matrices and tensors in a way that a mathematician would expect. 

<h3> For example, if you had two vectors: </h3>

- v1 = np.array([4, 5, 2, 7])
- v2 = np.array([2, 1, 3, 3])

And you add them together

v1 + v2

The result will be a ndarray where all the elements have been added together.

array([ 6, 6, 5, 10])

<h3> In contrast, if we had two Python Lists: </h3>

- list1 = [4, 5, 2, 7]
- list2 = [2, 1, 3, 3]
adding them together would just concatenate the lists.

list1 + list2
output: [4, 5, 2, 7, 2, 1, 3, 3]

Multiplying the two vectors together also results in an element by element operation:

v1 * v2

Gives us array([ 8, 5, 6, 21]) 

And for a Python List, this operation would not work at all.

list1 * list2 error!


<h2> Broadcasting </h2>

oftentimes you'll want to do some sort of operation between an array and a single number.

in mathematics, this single number is often called a scalar. 

For example, you might want to multiply every value in your NumPy array by 2:
![Adding](/tree/DATA_SCIENCE/Numpy/scalar.gif)

In order to achieve this result, NumPy will make the shape of the smaller array - our scalar - compatible with the larger array.

This is what the documentation refers to when it mentions the term "broadcasting".

The same rules about 'expanding' the smaller ndarray hold true for 2 or more dimensions. We can see this with a 2-Dimensional Array:

array_2d = np.array([[1, 2, 3, 4], 
                     [5, 6, 7, 8]])
                     
The scalar operates on an element by element basis.

array_2d + 1=

result = array_2d([[11, 12, 13, 14], 
                     [15, 16, 17, 18]])

### Manipulating Images as ndarrays:

Images are nothing other than a collection of pixels.

And each pixel is nothing other than value for a colour.

And any colour can be represented as a combination of red, green, and blue (RGB).


```python
img = misc.face()
plt.imshow(img)
```

```python
print(type(img)) # check the type of the image, which is a ndarray
print(img.shape)

print(img.ndim)
```

There are three matrices stacked on top of each other 
- one for the red values, 
- one for the green values 
- one for the blue values. 

Each matrix has a 768 rows and 1024 columns, which makes sense since 768x1024 is the resolution of the image.

### Challenge:

Now can you try and convert the image to black and white? All you need need to do is use a formula.

sRGB or black and white value range between 0 and 1, unlike RGB that range betwen 0 and 255

```python
sRGB_array = img/255
# Here NumPy will use broadcasting 
# to divide all the values in our ndarray by 255.
```

```python
# Next, we use matrix multiplication to multiply our two ndarrays together.
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
# These are the values given by the formula above
img_gray = sRGB_array @ grey_vals

# We can either multiply them together with the @ operator 
# or the .matmul() function.
img_gray = np.matmul(sRGB_array, grey_vals)
```

```python
# Finally, to show the image we use Matplotlib
plt.imshow(img_gray, cmap='gray')
```
